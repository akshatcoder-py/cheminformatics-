# **cheminformatics**

##**Load data**

```python
import  pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/dataprofessor/data/refs/heads/master/delaney_solubility_with_descriptors.csv')
df
```

# **Data peparations**

### Data separation as X and y

```python
x = df['logS']
x
```

```python
y = df.drop('logS', axis=1)
y
```

### Data splitting

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=100)
```

```python
x_train
```

```python
x_test
```

# **model building**

##**linear regression**

### **Training the model**

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(y_train, x_train)
```

### **Applying the model to make a prediction**

```python
y_lr_train_pred = lr.predict(y_train)
y_lr_test_pred = lr.predict(y_test)
```

```python
y_lr_train_pred
```

```python
y_lr_test_pred
```

###**Evaluate model performance**

```python
from sklearn.metrics import mean_squared_error, r2_score

lr_train_mse = mean_squared_error(x_train, y_lr_train_pred)
lr_train_r2 = r2_score(x_train, y_lr_train_pred)

lr_test_mse = mean_squared_error(x_test, y_lr_test_pred)
lr_test_r2 = r2_score(x_test, y_lr_test_pred)

print('LR MSE (Train): ', lr_train_mse)
print('LR R2 (Train): ', lr_train_r2)
print('LR MSE (Test): ', lr_test_mse)
print('LR R2 (Test): ', lr_test_r2)
```

```python
print("The Linear Regression model performance metrics (MSE and R2 for training and test sets) are now printed in the cell above.")
```

```python
import pandas as pd
lr_results = pd.DataFrame([['Linear regression', lr_train_mse, lr_train_r2, lr_test_mse, lr_test_r2]])
lr_results.columns = ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
```

```python
lr_results
```

# **Random forest**

###****Training the model****

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(max_depth=2, random_state=100)
rf.fit(y_train, x_train)
```

##***Applying the model to make a prediction***

```python
y_rf_train_pred = rf.predict(y_train)
y_rf_test_pred = rf.predict(y_test)
```

##**Evaluate model performance**

```python
from sklearn.metrics import mean_squared_error, r2_score

rf_train_mse = mean_squared_error(x_train, y_rf_train_pred)
rf_train_r2 = r2_score(x_train, y_rf_train_pred)

rf_test_mse = mean_squared_error(x_test, y_rf_test_pred)
rf_test_r2 = r2_score(x_test, y_rf_test_pred)
```

```python
rf_results = pd.DataFrame([['Random Forest', rf_train_mse, rf_train_r2, rf_test_mse, rf_test_r2]])
rf_results.columns = ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
rf_results
```

### **model comparision**

```python
df_models =  pd.concat([lr_results, rf_results], axis=0).reset_index()
df_models
```

# **Data visualization of prediction results**

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(5, 5))
plt.scatter(x=x_train, y=y_lr_train_pred, c = "#7CAE00", alpha=0.3)

z = np.polyfit(x_train, y_lr_train_pred, 1)
p = np.poly1d(z)

plt.plot(x_train, p(x_train), '#FF0000')
plt.title('cheminformatics graph\n')
plt.xlabel('Experimental')
plt.ylabel('Predicted logS')
```
